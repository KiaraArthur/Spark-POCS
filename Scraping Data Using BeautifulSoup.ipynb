```python
# -*- coding: utf-8 -*-
import MySQLdb
import pandas

db = MySQLdb.connect("localhost","root","prachi","recommender")

df2 = pandas.DataFrame(columns={'wordid', 'docid', 'word'})

cursor = db.cursor()

k=0
j=0

def unique_list(l):
    ulist = []
    [ulist.append(x) for x in l if x not in ulist]
    return ulist

for i in  range(1957):
    cursor.execute("SELECT article FROM docs WHERE docid="+str(i))
    data = cursor.fetchall()
    l=list(data)
    l=str(l).replace("[","").replace("]","").replace("(","").replace(")","").replace("'","").replace(","," ").replace("."," ").replace("/"," ").split()
    l=' '.join(unique_list(l)).split()
    j=len(l)+k
    d={'wordid' :range(k,j), 'docid' : i, 'word' : l}
    k=j
    df=pandas.DataFrame(d)
    df2=df2.append(df)    
    
    
#print df2   

from sqlalchemy import create_engine
database='recommender'
engine = create_engine('mysql+mysqldb://root:prachi@localhost/'+database)

df2.to_sql('d_words', engine, flavor='mysql', schema=None, if_exists='append', index=False, index_label={'wordid','docid','word'}, chunksize=None, dtype=None)
```
